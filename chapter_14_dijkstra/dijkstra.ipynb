### 데이크스트라  

특정 정점에서 모든 정점으로가는 최단거리 계산  

#### 조건
>무방향 그래프  
component 1개  
가중치는 양수  

#### 알고리즘  
0. 모든 가중치 INF  
1. 특정 정점 선택. 가중치 0으로 업데이트. 가중치과 정점 우선순위 큐에 삽입      
2. 우선순위 큐에서 최소 가중치 정점을 pop  
3. 선택된 정점과 연결된 정점의 가중치 합을 기존 가중치와 비교. 전자가 작으면 가중치 업데이트 및 우선순위 큐에 삽입  
4. 큐에 값이 없을 때 까지 반복 끝.
