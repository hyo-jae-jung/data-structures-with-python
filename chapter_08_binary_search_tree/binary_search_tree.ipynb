## Binary Search Tree(BST)  

Dictionary(<key,value>)를 구현 할 수 있는 자료구조는 BST와 hash table  
python dict는 hash table로 제작  


ADT  

Object:  
    - 유일한 키 값을 가진 노드 집합 -> TreeNode(key)  
    - TreeNode:  
        Object:   
            - key, left, right, parent  
        Operation:  
            - set: left, right, parent  
Operation:  
    BST.insert(key):  
        - 새로운 키 삽입  
        - 절차:  
            1. root 노드가 없으면 새로운 키가 root 노드  
            2. 기존 root가 있으면 root부터 비교 대상 키로 지정 후 새로운 키와 비교  
            2-1. 새로운 키가 비교 대상 키보다 작으면 새로운 비교 대상 키를 기존 비교   대상 키의 왼쪽 자식 노드로 변경  
            2-2. 새로운 키가 비교 대상 키보다 크면 새로운 비교 대상 키를 기존 비교 대상 키의 오른쪽 자식 노드로 변경  
            2-3. 새로운 비교 대상이 될 자식 노드가 없을 때까지 반복  
            3. 자식 노드가 없었던 부모 노드의 자식으로 새로운 키 삽입  
    BST.search(target) - > node:  
        - target을 키로 가지는 노드 반환  
        - 절차:  
            1. root 노드를 비교 대상 노드로 설정  
            2-1. target이 비교 대상 노드보다 작으면 비교 대상 노드의 왼쪽 자식 노드를 새로운 비교 대상 노드로 설정  
            2-2. target이 비교 대상 노드보다 크면 비교 대상 노드의 오른쪽 자식 노드를 새로운 비교 대상 노드로 설정  
            2-3. target과 비교 대상 노드가 같으면 해당 노드를 출력  
            2-4. 없으면 None 출력  
    BST.delete(target):  
        - target을 키로 가지는 노드 삭제  
        - 개념:삭제 대상 노드 기준, 자식 노드들로 자신의 위치도 채우고 BST 규칙 유지  
        - 절차:  
            1. target을 찾는다.  
            2-1. target의 자식이 없는 경우  
                2-1-1. target을 찾는 recursion의 결과를 None으로 설정  
            2-2. target의 자식이 왼쪽에만 있는 경우  
                2-2-1. target을 찾는 recursion의 결과를 target.left로 설정  
            2-3. target의 자식이 오른쪽에만 있는 경우  
                2-3-1. target을 찾는 recursion의 결과를 target.right로 설정  
            2-4. target의 자식이 둘인 경우  
                2-4-1. (변경 할 target 찾기)왼쪽 자식들 중 가장 큰 값의 자식을 target과 switching  
                2-4-2. (target 자식들 BST 규칙 적용)target의 왼쪽 자식으로 target과 변경됐던 reaf 노드를 찾는 recursion 진행 - 진행 결과 해당 reaf 노드는 None으로 설정  
            3. target찾기 바로 전 노드의 자식 노드로 연결  
            ※ delete는 recursion으로 구현돼서 코드를 봐야 정확히 이해 할 수 있음  
    BST.min(node) -> node:  
        - 기준 노드의 자식 노드 중 가장 작은 key 값을 가지는 node 출력  
        - 절차:
            1. 기준 노드의 왼쪽 자식 노드가 있으면 기준 노드를 왼쪽 자식 노드로 변경  
            2. 왼쪽 자식 노드가 없을 때 까지 1번 반복  
            3. 왼쪽 자식 노드가 없으면 해당 노드 선택  
    BST.max(node) -> node:  
        - 기준 노드의 자식 노드 중 가장 큰 key 값을 가지는 node 출력  
        - 절차:
            1. 기준 노드의 오른쪽 자식 노드가 있으면 기준 노드를 오른쪽 자식 노드로 변경  
            2. 오른쪽 자식 노드가 없을 때 까지 1번 반복  
            3. 오른쪽 자식 노드가 없으면 해당 노드 선택  
    BST.prev(cur) -> node:  
        - 정렬된 상태에서 선택된 노드의 바로 이전 노드 출력  
        - 절차:
            1. 선택된 노드의 왼쪽 자식 노드가 있으면 선택 후 종료  
            2. 왼쪽 자식 노드가 없으면 부모 노드 선택 후 자신이 부모 노드의 오른쪽 자식 노드인지 확인 맞으면 부모 노드 선택 후 종료   
            3. 아니면 부모 노드를 기준 노드로 변경. 2번 반복  
    BST.next(cur) -> node:  
        - 정렬된 상태에서 선택된 노드의 바로 이후 노드 출력  
        - 절차:
            1. 선택된 노드의 오른쪽 잣기 노드가 있으면 선택 후 종료  
            2. 오른쪽 자식 노드가 없으면 부모 노드 선택 후 자신이 부모 노드의 왼쪽 자식 노드인지 확인 맞으면 부모 노드 선택 후 종료  
            3. 아니면 부모 노드를 기준 노드로 변경. 2번 반복  

```python
class TreeNode:
    def __init__(self, key):
        self.__key=key
        self.__left=None
        self.__right=None
        self.__parent=None

    def __del__(self):
        print('key {} is deleted'.format(self.__key))

    @property
    def key(self):
        return self.__key

    @key.setter
    def key(self, key):
        self.__key=key

    @property
    def left(self):
        return self.__left

    @left.setter
    def left(self, left):
        self.__left=left

    @property
    def right(self):
        return self.__right

    @right.setter
    def right(self, right):
        self.__right=right

    @property
    def parent(self):
        return self.__parent

    @parent.setter
    def parent(self, p):
        self.__parent=p


class BST:
    def __init__(self):
        self.root=None

    def get_root(self):
        return self.root

    def preorder_traverse(self, cur, func):
        if not cur:
            return

        func(cur)
        self.preorder_traverse(cur.left, func)
        self.preorder_traverse(cur.right, func)

    def inorder_traverse(self, cur, func):
        if not cur:
            return

        self.inorder_traverse(cur.left, func)
        func(cur)
        self.inorder_traverse(cur.right, func)

    def __make_left(self, cur, left):
        cur.left=left
        if left:
            left.parent=cur

    def __make_right(self, cur, right):
        cur.right=right
        if right:
            right.parent=cur

    def insert(self, key):
        new_node=TreeNode(key)

        cur=self.root
        if not cur:
            self.root=new_node
            return

        while True:
            parent=cur
            if key < cur.key:
                cur=cur.left
                if not cur:
                    self.__make_left(parent, new_node)
                    return
            else:
                cur=cur.right
                if not cur:
                    self.__make_right(parent, new_node)
                    return 

    def search(self, target):
        cur=self.root
        while cur:
            if cur.key==target:
                return cur
            elif cur.key > target:
                cur=cur.left
            elif cur.key < target:
                cur=cur.right
        return cur

    def __delete_recursion(self, cur, target):
        if not cur:
            return None
        elif target < cur.key:
            new_left=self.__delete_recursion(cur.left, target)
            self.__make_left(cur, new_left)
        elif target > cur.key:
            new_right=self.__delete_recursion(cur.right, target)
            self.__make_right(cur, new_right)
        else:
            if not cur.left and not cur.right:
                cur=None
            elif not cur.right:
                cur=cur.left
            elif not cur.left:
                cur=cur.right
            else:
                replace=cur.left
                replace=self.max(replace)
                cur.key, replace.key=replace.key, cur.key
                new_left=self.__delete_recursion(cur.left, replace.key)
                self.__make_left(cur, new_left)
        return cur

    def delete(self, target):
        new_root=self.__delete_recursion(self.root, target)
        self.root=new_root

    def min(self, cur):
        while cur.left != None:
            cur=cur.left
        return cur 

    def max(self, cur):
        while cur.right != None:
            cur=cur.right
        return cur

    def prev(self, cur):
        if cur.left:
            return self.max(cur.left)
        
        parent = cur.parent
        while parent and cur==parent.left:
            cur=parent
            parent=parent.parent
        
        return parent

    def next(self, cur):
        if cur.right:
            return self.min(cur.right)        

        parent = cur.parent
        while parent and cur==parent.right:
            cur=parent
            parent=parent.parent
        
        return parent      

if __name__=="__main__":
    print('*'*100)
    bst=BST()

    bst.insert(6)
    bst.insert(3)
    bst.insert(2)
    bst.insert(4)
    bst.insert(5)
    bst.insert(8)
    bst.insert(10)
    bst.insert(9)
    bst.insert(11)

    f=lambda x: print(x.key, end='  ')

    #bst.preorder_traverse(bst.get_root(), f)
    bst.inorder_traverse(bst.get_root(), f)
    print()
    
    searched_node = bst.search(8)
    if searched_node:
        print(f'searched key : {searched_node.key}')
        
        prev_node = bst.prev(searched_node)
        if prev_node:
            print(f'prev key : {prev_node.key}')
        else:
            print(f'this is the first key of the BST')

        next_node = bst.next(searched_node)
        if next_node:
            print(f'next key : {next_node.key}')
        else:
            print(f'this is the last key of the BST')
    else:
        print('there is no such key')
    print()


    print (f'MIN(bst) : {bst.min(bst.get_root()).key}')
    print(f'MAX(bst) : {bst.max(bst.get_root()).key}')

    #bst.delete(9)
    #bst.delete(8)
    bst.delete(6)

    #print(bst.delete(15))

    bst.preorder_traverse(bst.get_root(), f)
    # bst.inorder_traverse(bst.get_root(), f)
    print()
    print('*'*100)
```
